## Lab: RNN with Forecasting

### Wind Turbine Power Prediction with RNN

This lab uses a Recurrent Neural Network to predict wind turbine power output.
* [Data link](https://www.kaggle.com/code/ahmedfathygwely/wind-turbine-dataset-machine-learning-rnn-times/input)

### Step 0: Load libraries & Data

```python
# Step 0: Load libraries & Data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import math

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Load the data
df = pd.read_csv('T1.csv')
print(df.columns.tolist())
```

```python
# Display basic information
print(df.shape)
df.info()
print(df.describe())
print(df.head())
```

### Step 1: Prepare the data

```python
# Step 1: Prepare the data
df.columns = df.columns.str.strip()
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format="%d %m %Y %H:%M") 
df.isnull().sum() 
df.describe()

plt.figure(figsize=(8, 5))
plt.scatter(df['Wind Speed (m/s)'], df['LV ActivePower (kW)'], alpha=0.3)
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('LV ActivePower (kW)')
plt.title('Wind Speed vs. LV ActivePower')
plt.grid(True)
plt.show()

selected_features = ['LV ActivePower (kW)', 'Wind Speed (m/s)', 'Theoretical_Power_Curve (KWh)']
df_selected = df[selected_features].copy()
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(df_selected.values)
```

### Step 2: Create sequences

```python
# Step 2: Create sequences
print(df['Date/Time'].head(10)) 
interval = df.index.to_series().diff().dropna()
print(interval.value_counts())
seq_length = 144 # 24hrs x 6 samples(every 10 min)

def create_sequences(data, seq_length):
    """
    Turns a 2D array (samples x features) into a 3D array of sliding windows (samples x seq_length x features)
    and a 2D array of next-step targets (samples x features).
    
    Parameters:
    -----------
    data : np.ndarray
        Scaled dataset, shape = (num_samples, num_features)
    seq_length : int
        Length of each input sequence (number of time steps)
    
    Returns:
    --------
    X : np.ndarray
        Input sequences, shape = (num_sequences, seq_length, num_features)
    y : np.ndarray
        Corresponding next values, shape = (num_sequences, num_features) or (num_sequences,) if single target
    """
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i : i + seq_length])
        y.append(data[i + seq_length])  # Predict all features
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_data, seq_length)

print(X.shape)
print(y.shape)
```

### Step 3: Split the data

```python
# Step 3: Split the data
train_size = int(len(X) * 0.8)

X_train, X_test = X[:train_size], X[train_size:]

y_train, y_test = y[:train_size], y[train_size:]

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

### Step 4: Build the RNN Model

```python
# Step 4: Build the RNN Model
model = Sequential()
model.add(LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(32, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1))  # 1 unit for predicting a single value

model.compile(optimizer='adam', loss='mean_squared_error')

# Print model summary
model.summary()
```

### Step 5: Train the model

```python
# Step 5: Train the model
# Stops training if val_loss doesn't improve for 5 epochs
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Saves the best model (lowest val_loss)
checkpoint = ModelCheckpoint(
    filepath='best_model.h5',
    monitor='val_loss',
    save_best_only=True,
    verbose=1
)

history = model.fit(
    X_train, y_train[:, 0],                 # Only LV ActivePower
    epochs=100,                            
    batch_size=32,
    validation_data=(X_test, y_test[:, 0]),
    callbacks=[early_stop, checkpoint],
    verbose=1
)

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss')
plt.legend()
plt.show()
```

### Step 6: Evaluate on the test data and visualize the results

```python
# Step 6: Evaluate and visualize
# Predict on test set
y_pred_scaled = model.predict(X_test)

# Prepare an array to hold inverse scaled results (must match scaler's expected input shape)
y_pred_full = np.zeros((len(y_pred_scaled), scaled_data.shape[1]))
y_test_full = np.zeros((len(y_test), scaled_data.shape[1]))

# Put predictions and actuals in the right column (LV ActivePower is column 0)
y_pred_full[:, 0] = y_pred_scaled.flatten()
y_test_full[:, 0] = y_test[:, 0]

# Inverse transform
y_pred_inv = scaler.inverse_transform(y_pred_full)[:, 0]
y_test_inv = scaler.inverse_transform(y_test_full)[:, 0]

rmse = math.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
r2 = r2_score(y_test_inv, y_pred_inv)

print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.4f}")
```

```python

# Time-series plot: first 500 points for clarity
plt.figure(figsize=(14, 5))
plt.plot(y_test_inv[:500], label='Actual')
plt.plot(y_pred_inv[:500], label='Predicted')
plt.legend()
plt.title('LV ActivePower (kW): Actual vs Predicted (First 500 points)')
plt.xlabel('Time step')
plt.ylabel('Power Output (kW)')
plt.show()

# Scatter plot: All test samples
plt.figure(figsize=(6, 6))
plt.scatter(y_test_inv, y_pred_inv, alpha=0.3)
plt.xlabel('Actual Power Output (kW)')
plt.ylabel('Predicted Power Output (kW)')
plt.title('Predicted vs Actual: LV ActivePower (kW)')
plt.plot([y_test_inv.min(), y_test_inv.max()], [y_test_inv.min(), y_test_inv.max()], 'r--')
plt.show()

plt.figure(figsize=(14, 5))
start, end = 1000, 1200  # Change this window as needed
plt.plot(range(start, end), y_test_inv[start:end], label='Actual')
plt.plot(range(start, end), y_pred_inv[start:end], label='Predicted')
plt.legend()
plt.title(f'Zoomed View: Actual vs Predicted (Samples {start}–{end})')
plt.xlabel('Time step')
plt.ylabel('Power Output (kW)')
plt.show()
```
