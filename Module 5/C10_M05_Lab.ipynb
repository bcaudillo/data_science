### RNN with Attention (Possible solution)

```python
# -*- coding: utf-8 -*-
"""
Wind Turbine Power Prediction with RNNs + Attention

This lab uses an Attention-enhanced Recurrent Neural Network to predict wind turbine power output.
"""

# Step 0: Load libraries & Data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, Layer
from tensorflow.keras.callbacks import EarlyStopping
import math

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Step 1: Prepare the Data
df = pd.read_csv('T1.csv')
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format="%d %m %Y %H:%M", dayfirst=True)
df = df.dropna()

# Quick EDA
print(df.describe())
plt.scatter(df['Wind Speed (m/s)'], df['LV ActivePower (kW)'])
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('LV ActivePower (kW)')
plt.title('Wind Speed vs Active Power')
plt.show()

```

```python
# Step 2: Create sequences
features = ['LV ActivePower (kW)', 'Wind Speed (m/s)', 'Theoretical_Power_Curve (KWh)']
data=df[features].values

#Normalize
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

print(df['Date/Time'].head(10)) 
interval = df.index.to_series().diff().dropna()
print(interval.value_counts())
seq_length = 24 # 24hrs x 6 samples(every 10 min)

def create_sequences(data, seq_length):
    """
    Turns a 2D array (samples x features) into a 3D array of sliding windows (samples x seq_length x features)
    and a 2D array of next-step targets (samples x features).
    
    Parameters:
    -----------
    data : np.ndarray
        Scaled dataset, shape = (num_samples, num_features)
    seq_length : int
        Length of each input sequence (number of time steps)
    
    Returns:
    --------
    X : np.ndarray
        Input sequences, shape = (num_sequences, seq_length, num_features)
    y : np.ndarray
        Corresponding next values, shape = (num_sequences, num_features) or (num_sequences,) if single target
    """
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length][0])  # Predict LV ActivePower
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_data, seq_length)

print(X.shape)
print(y.shape)
```

```python
# Step 3: Split the data
train_size = int(len(X) * 0.8)

X_train, X_test = X[:train_size], X[train_size:]

y_train, y_test = y[:train_size], y[train_size:]

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Step 4: Define a Custom Attention Layer
class AttentionLayer(Layer):
    def __init__(self, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)
    def build(self, input_shape):
        self.W = self.add_weight(name="att_weight", shape=(input_shape[-1], input_shape[-1]), initializer="normal")
        self.b = self.add_weight(name="att_bias", shape=(input_shape[-1],), initializer="zeros")
        super(AttentionLayer, self).build(input_shape)
    def call(self, inputs):
        score = tf.tanh(tf.matmul(inputs, self.W) + self.b)
        attention_weights = tf.nn.softmax(score, axis=1)
        context_vector = attention_weights * inputs
        context_vector = tf.reduce_sum(context_vector, axis=1)
        return context_vector

```

```python
# Step 5: Build the RNN Model with Attention
inputs = Input(shape=(seq_length, len(features)))
x = LSTM(64, return_sequences=True)(inputs)
x = Dropout(0.2)(x)
x = LSTM(32, return_sequences=True)(x)
x = Dropout(0.2)(x)
context_vector = AttentionLayer()(x)
output = Dense(1)(context_vector)
model = Model(inputs, output)
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
# Step 6: Compile and Train the Model
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop]
)

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss')
plt.legend()
plt.show()

# Step 7: Evaluate and Visualize the Results
y_pred = model.predict(X_test)
# Inverse transform predictions and targets for metrics
y_test_rescaled = scaler.inverse_transform(np.concatenate([y_test.reshape(-1,1), np.zeros((y_test.shape[0], data.shape[1]-1))], axis=1))[:,0]
y_pred_rescaled = scaler.inverse_transform(np.concatenate([y_pred, np.zeros((y_pred.shape[0], data.shape[1]-1))], axis=1))[:,0]

mse = mean_squared_error(y_test_rescaled, y_pred_rescaled)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_rescaled, y_pred_rescaled)
print(f"RMSE: {rmse:.2f}")
print(f"RÂ²: {r2:.4f}")
```

```python
# Visualize attention weights 
plt.figure(figsize=(15, 5))
plt.plot(y_test_rescaled, label='Actual', linewidth=2, color='blue')
plt.plot(y_pred_rescaled, label='Predicted', linewidth=2, color='orange', alpha=0.7)
plt.xlabel('Time Step')
plt.ylabel('Power Output (kW)')
plt.title('Actual vs Predicted Power Output')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 6))
plt.scatter(y_test_rescaled, y_pred_rescaled, alpha=0.5)
plt.plot([min(y_test_rescaled), max(y_test_rescaled)],
         [min(y_test_rescaled), max(y_test_rescaled)], 'r--', linewidth=2, label='Ideal Fit')
plt.xlabel('Actual Power Output (kW)')
plt.ylabel('Predicted Power Output (kW)')
plt.title('Actual vs Predicted Scatter Plot')
plt.legend()
plt.tight_layout()
plt.show()
```
