```python
# Generated Code:
import pandas as pd
import matplotlib.pyplot as plt

def analyze_user_engagement(logs_df):
    # Data validation
    if logs_df.isnull().any().any():
        raise ValueError("DataFrame contains missing values.")
    
    if (logs_df['duration'] < 0).any():
        raise ValueError("Duration contains negative values.")
        
    if (logs_df['page_views'] <= 0).any():
        raise ValueError("Page views must be positive integers.")
    
    # Aggregate session-level metrics
    session_data = logs_df.groupby('session_id').agg({
        'duration': 'sum',
        'page_views': 'sum'
    }).reset_index()
    
    # Calculate bounce rate: sessions with exactly 1 page view
    bounce_rate = (session_data['page_views'] == 1).mean()
    
    # Average session time
    avg_session_time = session_data['duration'].mean()
    
    # Pages per session
    pages_per_session = session_data['page_views'].mean()
    
    metrics = {
        'bounce_rate': round(bounce_rate * 100, 2),
        'avg_session_time': round(avg_session_time, 2),
        'pages_per_session': round(pages_per_session, 2)
    }
    
    # Device-level metrics
    device_metrics = logs_df.groupby('device_type').agg({
        'session_id': 'count',
        'duration': 'mean',
        'page_views': 'sum'
    }).rename(columns={'session_id': 'session_count', 'duration': 'avg_duration'})
    
    # Visual Summary
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))

    # Average session time by device
    device_metrics['avg_duration'].plot(kind='bar', ax=ax[0])
    ax[0].set_title('Average Session Time by Device')
    ax[0].set_ylabel('Seconds')

    # Simulated bounce rate by device (for illustration)
    session_bounce = logs_df.groupby(['device_type', 'session_id'])['page_views'].sum().reset_index()
    bounce_by_device = session_bounce.groupby('device_type').apply(lambda x: (x['page_views'] == 1).mean() * 100)

    bounce_by_device.plot(kind='bar', ax=ax[1], color='orange')
    ax[1].set_title('Bounce Rate by Device')
    ax[1].set_ylabel('Percentage')

    plt.tight_layout()
    plt.show()
    
    return metrics, device_metrics
```

```python
# Final Solution:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

"""
README: User Engagement Analysis Script

This script processes user session log data from a fitness tracking app
and generates engagement metrics and visual summaries to support stakeholder insights.

Fixes and Additions:
- Added data validation to prevent misleading metrics
- Corrected bounce rate calculation based on session-level aggregation
- Included new statistics such as total sessions and engagement distributions
- Replaced bar plots with scatter plots and heatmaps for richer exploration
- Added detailed inline comments for maintainability
"""

def analyze_user_engagement(logs_df):
    # --- Data Validation ---
    if logs_df.isnull().any().any():
        raise ValueError("The dataset contains missing values. Please clean the data before analysis.")
    
    if (logs_df['duration'] < 0).any():
        raise ValueError("Duration column has negative values. All durations should be non-negative.")
    
    if (logs_df['page_views'] <= 0).any():
        raise ValueError("Page views must be greater than zero for all sessions.")
    
    # --- Session-Level Aggregation ---
    session_data = logs_df.groupby('session_id').agg({
        'duration': 'sum',
        'page_views': 'sum'
    }).reset_index()
    
    # Bounce Rate = sessions with exactly 1 page view / total sessions
    bounce_rate = (session_data['page_views'] == 1).mean() * 100

    # Compute key metrics
    avg_session_time = session_data['duration'].mean()
    pages_per_session = session_data['page_views'].mean()

    metrics = {
        'total_sessions': session_data.shape[0],
        'bounce_rate_percent': round(bounce_rate, 2),
        'avg_session_time_sec': round(avg_session_time, 2),
        'pages_per_session': round(pages_per_session, 2)
    }

    # --- Device-Level Metrics ---
    device_metrics = logs_df.groupby('device_type').agg({
        'session_id': 'count',
        'duration': 'mean',
        'page_views': 'sum'
    }).rename(columns={
        'session_id': 'session_count',
        'duration': 'avg_duration_sec'
    })

    # --- Visualizations ---
    plt.figure(figsize=(16, 12))

    # Scatter Plot: Duration vs Page Views
    plt.subplot(2, 2, 1)
    sns.scatterplot(data=session_data, x='duration', y='page_views', alpha=0.6)
    plt.title('Page Views vs Session Duration')
    plt.xlabel('Duration (seconds)')
    plt.ylabel('Page Views')
    plt.grid(True)

    # Heatmap: Correlation Between Numeric Fields
    plt.subplot(2, 2, 2)
    corr_matrix = session_data[['duration', 'page_views']].corr()
    sns.heatmap(corr_matrix, annot=True, cmap='Blues', fmt='.2f', square=True)
    plt.title('Correlation Heatmap')

    # Distribution Plot: Session Duration
    plt.subplot(2, 2, 3)
    sns.histplot(session_data['duration'], bins=15, kde=True, color='green')
    plt.title('Distribution of Session Durations')
    plt.xlabel('Duration (seconds)')
    plt.ylabel('Frequency')

    # Distribution Plot: Page Views per Session
    plt.subplot(2, 2, 4)
    sns.histplot(session_data['page_views'], bins=15, kde=True, color='purple')
    plt.title('Distribution of Page Views per Session')
    plt.xlabel('Page Views')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

    return metrics, device_metrics
```

```python
"""
#Your prompt
I need help writing a SQL query to support the product team. We're trying to segment customers for a new feature rollout. I want to start by identifying active users, meaning people who have logged in within the last 30 days. Then I want to filter that group down to the top 20 percent of spenders based on total transaction amount. After that, I need to pull in their user preferences so we can look at trends like communication style or interface choices. The database has three tables: user_activity, transactions, and user_preferences. Can you help me write a clean query that handles all of this?
"""
```

```python
#Generated Code
WITH active_users AS (
    SELECT user_id
    FROM user_activity
    WHERE last_login_date >= CURRENT_DATE - INTERVAL '30 days'
),

user_spending AS (
    SELECT
        user_id,
        SUM(amount) AS total_spent
    FROM transactions
    GROUP BY user_id
),

spending_with_percentile AS (
    SELECT
        user_id,
        total_spent,
        NTILE(5) OVER (ORDER BY total_spent DESC) AS spending_tier
    FROM user_spending
),

high_value_active_users AS (
    SELECT s.user_id
    FROM spending_with_percentile s
    JOIN active_users a ON s.user_id = a.user_id
    WHERE s.spending_tier = 1
)

SELECT
    u.user_id,
    p.communication_preference,
    p.interface_theme,
    p.notification_settings
FROM high_value_active_users u
JOIN user_preferences p ON u.user_id = p.user_id;
```

```python
"""
README: High-Value Active User Segmentation for Feature Rollout

Purpose:
This script segments users for a targeted product rollout by identifying customers who are:
1. Active (logged in within the last 30 days)
2. High-value (top 20 percent by total spending)

The output includes user preferences such as communication preference, interface theme, and notification settings.
Supporting queries provide insights into platform trends, account-type spend behavior, and preference breakdowns.

Tables Used:
- user_activity: user login and feature usage data
- transactions: individual transaction records with platform and amount
- user_preferences: user-selected communication and display settings

Intended Stakeholders:
- Product team: for rollout planning and feature targeting
- Marketing: for personalization strategies
- Data and engineering teams: for segmentation support and validation
"""

#Step 1: Identify active users (logged in within the last 30 days)
WITH active_users AS (
    SELECT 
        user_id
    FROM user_activity
    WHERE last_login_date >= CURRENT_DATE - INTERVAL '30 days'
),

#Step 2: Calculate total spend per user
user_spending AS (
    SELECT
        user_id,
        SUM(amount) AS total_spent
    FROM transactions
    GROUP BY user_id
),

#Step 3: Rank users into 5 spending tiers using NTILE
#Tier 1 represents the top 20 percent of spenders
spending_with_percentile AS (
    SELECT
        user_id,
        total_spent,
        NTILE(5) OVER (ORDER BY total_spent DESC) AS spending_tier
    FROM user_spending
),

#Step 4: Filter for users who are both active and high-value (top tier)
high_value_active_users AS (
    SELECT s.user_id
    FROM spending_with_percentile s
    JOIN active_users a ON s.user_id = a.user_id
    WHERE s.spending_tier = 1
)

#Final Output: Preferences for high-value, active users
SELECT
    u.user_id AS high_value_user_id,
    p.communication_preference,
    p.interface_theme,
    p.notification_settings
FROM high_value_active_users u
JOIN user_preferences p ON u.user_id = p.user_id;



#Supporting Query 1: Platform trends among top spenders
#This shows where revenue is coming from (mobile, web, etc.)
SELECT
    t.platform,
    COUNT(DISTINCT t.user_id) AS user_count,
    SUM(t.amount) AS total_revenue
FROM transactions t
JOIN (
    SELECT user_id
    FROM spending_with_percentile
    WHERE spending_tier = 1
) high_spenders ON t.user_id = high_spenders.user_id
GROUP BY t.platform
ORDER BY total_revenue DESC;



#Supporting Query 2: Average spend per user by account type
#Helps compare engagement between free vs premium users
SELECT
    a.account_type,
    COUNT(DISTINCT t.user_id) AS user_count,
    ROUND(SUM(t.amount) / COUNT(DISTINCT t.user_id), 2) AS avg_spend_per_user
FROM transactions t
JOIN user_activity a ON t.user_id = a.user_id
GROUP BY a.account_type;



#Supporting Query 3: Communication preference distribution
#Helps identify how high-value users prefer to receive messages
SELECT
    communication_preference,
    COUNT(*) AS user_count
FROM user_preferences
WHERE user_id IN (
    SELECT user_id
    FROM high_value_active_users
)
GROUP BY communication_preference
ORDER BY user_count DESC;
```

```python
"""
How were percentile calculations handled?
The query used the NTILE(5) function to break users into five spending tiers. That made it easy to isolate the top 20 percent of spenders by assigning them to tier 1. It’s a simple way to calculate percentiles without having to manually define thresholds, and it works well when the goal is to create evenly sized groups based on a ranking.

What approaches to date filtering were used?
For date filtering, we focused on users who logged in within the last 30 days. That was done with a filter using last_login_date >= CURRENT_DATE - INTERVAL '30 days'. This kind of relative date filter helps keep the logic dynamic and ensures that we’re always working with the most recent activity, regardless of when the query runs.

How was the query optimized?
The query used common table expressions (CTEs) to break things up into clear, reusable steps. That made it easier to read and maintain. Each part of the logic—finding active users, calculating spending, ranking users by percentile, and filtering for high-value users—was separated into its own CTE. This also helped avoid repeating logic or nesting too many subqueries, which can make things harder to follow and debug.
"""
```
