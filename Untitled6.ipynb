```python
# Required Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Datasets
wine = pd.read_csv('Wine Data.csv')
chickwts = pd.read_csv('Chick Weight Data.csv')
us_arrests = pd.read_csv('Arrests Data.csv')
```

```python
# Exploratory Data Analysis
print("Wine Dataset Info:")
print(wine.info())
print(wine.describe(), "\n")
```

```python
print("Chickwts Dataset Info:")
print(chickwts.info())
print(chickwts.describe(), "\n")
```

```python
print("USArrests Dataset Info:")
print(us_arrests.info())
print(us_arrests.describe(), "\n")
```

```python
# ========== Data Cleaning and Preprocessing ==========
# Wine Dataset
wine.dropna(inplace=True)

# Separate features and target
wine_features = wine.drop('target', axis=1)
wine_target = LabelEncoder().fit_transform(wine['target'])

# Standardize features
scaler_wine = StandardScaler()
wine_scaled = scaler_wine.fit_transform(wine_features)

#Chickwts Dataset Prep
# Drop non-numeric column
chickwts_numeric = chickwts.drop(columns=['feed_name'])

# Ensure no invalid weight values
chickwts_clean = chickwts_numeric[chickwts_numeric['weight'] > 0]

# Standardize all features
scaler_chick = StandardScaler()
chickwts_scaled = scaler_chick.fit_transform(chickwts_clean)

#US Arrests Dataset Prep
us_arrests_clean = us_arrests.drop(columns=['Unnamed: 0'])
scaler_us = StandardScaler()
us_arrests_scaled = scaler_us.fit_transform(us_arrests_clean)
```

```python
#PCA for Wine Dataset
pca_wine = PCA(n_components=0.95)
wine_pca = pca_wine.fit_transform(wine_scaled)

pca = PCA(n_components=2)
us_pca = pca.fit_transform(us_scaled)


k_values = list(range(2, 10))
silhouette_scores_kmeans = []
silhouette_scores_gmm = []


# GridSearchCV for k-NN
param_grid = {
    'n_neighbors': list(range(1, 21)),
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=5)
grid_search.fit(wine_pca, wine_target)

# Evaluate k-NN
wine_pred = grid_search.predict(wine_pca)
print(classification_report(wine_target, wine_pred))
print("Accuracy:", accuracy_score(wine_target, wine_pred))
```

```python
# Cosine Similarity for Chickwts
pca_chick = PCA(n_components=1)
chick_pca = pca_chick.fit_transform(chickwts_scaled)
similarity_matrix = cosine_similarity(chick_pca)
print("Cosine Similarity:\n", similarity_matrix)
```

```python
# Clustering on USArrests 
# Feature selection (top 3 by variance)
feature_variances = np.var(us_arrests_scaled, axis=0)
top3_idx = np.argsort(feature_variances)[-3:]
us_selected = us_arrests_scaled[:, top3_idx]

# PCA to 2 components
pca_us = PCA(n_components=2)
us_pca = pca_us.fit_transform(us_selected)
```

```python
# Elbow Method
inertias = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(us_pca)
    inertias.append(kmeans.inertia_)

# Plot Elbow
plt.figure(figsize=(6,4))
plt.plot(range(1, 10), inertias, marker='o')
plt.title('Elbow Method (KMeans)')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.tight_layout()
plt.show()
```

```python
# BIC for GMM
bic_scores = []
for k in range(1, 10):
    gmm = GaussianMixture(n_components=k, random_state=42)
    gmm.fit(us_pca)
    bic_scores.append(gmm.bic(us_pca))

# Plot BIC
plt.figure(figsize=(6,4))
plt.plot(range(1, 10), bic_scores, marker='x')
plt.title('BIC Scores (GMM)')
plt.xlabel('K')
plt.ylabel('BIC')
plt.tight_layout()
plt.show()

n_components = range(1, 10)
aic_scores = []
```

```python
# Fit GMM models and compute AIC for each
for n in n_components:
    gmm = GaussianMixture(n_components=n, random_state=42)
    gmm.fit(us_arrests_scaled)
    aic_scores.append(gmm.aic(us_arrests_scaled))

# Plotting AIC scores
plt.figure(figsize=(8, 5))
plt.plot(n_components, aic_scores, marker='o')
plt.title('AIC vs Number of GMM Components')
plt.xlabel('Number of Clusters')
plt.ylabel('AIC Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Optimal clustering with K=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
us_arrests['KMeans_Cluster'] = kmeans.fit_predict(us_pca)

gmm = GaussianMixture(n_components=3, random_state=42)
us_arrests['GMM_Cluster'] = gmm.fit_predict(us_pca)
```

```python
# Visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.scatterplot(x=us_pca[:, 0], y=us_pca[:, 1], hue=us_arrests['KMeans_Cluster'], ax=axes[0], palette='Set2')
axes[0].set_title('K-Means Clustering')
sns.scatterplot(x=us_pca[:, 0], y=us_pca[:, 1], hue=us_arrests['GMM_Cluster'], ax=axes[1], palette='Set3')
axes[1].set_title('GMM Clustering')
plt.tight_layout()
plt.show()
```
